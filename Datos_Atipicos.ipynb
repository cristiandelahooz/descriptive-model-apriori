```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Leer el archivo CSV
df = pd.read_csv("Asociacion.csv")

# Crear el boxplot para una columna específica
plt.figure(figsize=(8, 6))
sns.boxplot(x=df["impresiones"], color="lightblue")
plt.title("Boxplot de impresiones")
plt.xlabel("Valores")
plt.show()

# Calcular los outliers utilizando el método IQR
Q1 = df["impresiones"].quantile(0.25)
Q3 = df["impresiones"].quantile(0.75)
IQR = Q3 - Q1
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# Filtrar los outliers
outliers = df[
    (df["impresiones"] < limite_inferior) | (df["impresiones"] > limite_superior)
]
print("Outliers encontrados:")
print(outliers)
```
